```python
#Anabaena Catenula (Algae Growth)

def Evolve(myString): #Executes reproduction rules
    if myString == 'A':
        return 'B'
    elif myString == 'B':
        return 'C'
    elif myString == 'C':
        return 'D'
    elif myString == 'D':
        return 'E'
    elif myString == 'E':
        return 'F'
    elif myString == 'F':
        return 'AB'

generations = 50
Lsystem = 'A' #axioma inicial
for i in range(generations):
    temp = ''
    print(Lsystem)
    
    #go through each letter of the Lsystem and evolve it
    for j in range(len(Lsystem)):
        temp = temp + Evolve(Lsystem[j])

    #set and print
    Lsystem = temp


print('number of algae')
print(len(Lsystem))
```
